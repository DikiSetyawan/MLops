```python
import pandas as pd 
df = pd.read_csv(r'D:\project\porto\data\data1.csv')
df['time'] = pd.to_datetime(df['time'])
df.head()
```

```python
jakarta = df[df['City'].isin(['Jakarta'])]
jakarta.head(1000)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_features = ['Temperature (C)', 'Humidity (%)', 'Wind Speed (m/s)']
for col in numeric_features:
    plt.hist(jakarta[col], bins=20)  # Access the column using df[col]
    plt.title(col)  # Add a title to the histogram to know which feature it represents
    plt.show()
```

```python
import matplotlib.pyplot as plt

numeric_features = ['Temperature (C)', 'Humidity (%)', 'Wind Speed (m/s)']

for col in numeric_features:
    plt.scatter(jakarta['time'], jakarta[col])
    plt.xlabel('Time')
    plt.ylabel(col)
    plt.title(f'Scatter plot of {col} against Time')
    plt.show()
```
